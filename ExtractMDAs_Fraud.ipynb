# PreProcess MDA Section Test

```python
from autocorrect import spell
```

```python
import nltk
from nltk.stem import WordNetLemmatizer 
#is based on The Porter Stemming Algorithm
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
from nltk.tokenize import sent_tokenize
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
def cleanString(mda_extract):
    stopWords = set(stopwords.words('english'))
    lemmatizer = WordNetLemmatizer()
    returnString = ""
    sentence_token = nltk.sent_tokenize(mda_extract)
    idx_list = []
    for j in range(len(sentence_token)):
        single_sentence = nltk.word_tokenize(sentence_token[j])
        sentences_filtered = [(idx,lemmatizer.lemmatize(spell(w).lower())) for idx,w in enumerate(single_sentence) 
                              if w.lower() not in stopWords and w.isalnum() and not w.isdigit()]
        idx_list.append([x[0] for x in sentences_filtered])
        word_list = [x[1] for x in sentences_filtered]
        returnString = returnString + ' '.join(word_list) + ' . '
    
    return returnString, idx_list

```

```python
def total_word_count(text):
    word_token=nltk.tokenize.word_tokenize(text)
    return len(word_token)
#Counting total words

def total_sentence_count(text):
    sentence_token = nltk.tokenize.sent_tokenize(text)
    return len(sentence_token)
```
